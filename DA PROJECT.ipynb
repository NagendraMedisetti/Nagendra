```python
label=['yes', 'no']
Y=df[df['freepoor']=='yes']
N=df[df['freepoor']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x, labels=label)
plt.title("% of people getting govt health Insurance due to low income ")
plt.show()
#% of people having private Insurance
Y = df[df['private']=='yes'] 
N=df[df[ 'private']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x, labels=label)
plt.title("% of people having private health Insurance ")
plt.show()
# % of people getting govt Insurance due to old age, disabil disabili or veteran status
Y=df[df['freerepat']=='yes']
N=df[df['freerepat']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x, labels=label)
plt.title("% of people getting govt health Insurance due to old age, disability or veteran status ")
plt.show()
```

```python
db= df.groupby('gender')['reduced'].sum().to_frame().reset_index()
# Creating the bar chart
plt.barh(db['gender'], db['reduced'], color = ['cornflowerblue', 'lightseagreen'])
#Adding the aesthetics
plt.title('Bar Chart')
plt.xlabel('reduced activity')
plt.ylabel('gender')
#Show the plot
plt.show()
```

```python
unique_income = df['income'].unique()
unique_health = df['health'].unique()
unique_income.sort()
unique_health.sort()
income_counts = df['income'].value_counts()
health_counts = df['health'].value_counts()
plt.bar(unique_income, income_counts, color='blue', label='Income')
plt.bar(unique_health, health_counts, color='green', label='Health')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.title('Comparison between Income and Health')
plt.legend()
plt.show()
```

```python
gender_counts = df['gender'].value_counts()
colors = ['blue', 'pink']
labels = ['Male', 'Female']
sizes = [gender_counts.get('male', 0), gender_counts.get('female', 0)]

fig, ax = plt.subplots()
ax.barh(labels, sizes, color=colors)
info_text = f'Male: {gender_counts.get("male", 0)}\nFemale: {gender_counts.get("female", 0)}'
ax.text(0.05, 0.95, info_text, fontsize=5, transform=ax.transAxes, bbox=dict(facecolor='white', alpha=0.5))
ax.set_title('Comparison of Males and Females Visited')
plt.show()
```
